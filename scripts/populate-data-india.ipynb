# Create a district id for each district

```python
cen.rename(columns={'State': 'state_id', 'District': 'district_id', 'Name': 'district', 'TOT_P': 'total_persons', 'TOT_M': 'total_males', 'TOT_F': 'total_females', 'P_06': 'children_persons', 'M_06': 'children_males', 'F_06': 'children_females'}, inplace=True)
cen = cen[[
    'district_id', 'district', 'state_id', 'total_persons', 'total_males', 'total_females', 'children_persons', 'children_males', 'children_females'
]]
cen
#cen[cen['State name'] == 'MAHARASHTRA' ].sort_values(by='District name')
```

```python
# remove the "rural" and "urban" parts of each district
indicestodrop = []
for index, row in cen.iterrows():
    if index % 3 == 0:
        continue
    indicestodrop.append(index)
cen.drop(indicestodrop, inplace = True)
cen.to_excel('2011-IndiaStateDist-0000.xlsx', index=False)
```

```python
# Combine DADRA & NAGAR HAVELI and DAMAN & DIU
ret = []
ind = []
for index, row in cen.iterrows():
    if row['district'] in ['DADRA & NAGAR HAVELI', 'DAMAN & DIU']:
        ret.append(list(row))
        ind.append(index)
print(ret)
ret2 = np.array(ret[0][2:]) + np.array(ret[1][2:])
print(ret2)
cen.drop(ind, inplace=True)
ret2 = list(ret2)
ret2 = [0, 'DADRA & NAGAR HAVELI & DAMAN & DIU'] + ret2
print(ret2)
cen.loc[len(cen)] = ret2
cen.to_csv('2011-IndiaStateDist-0000.xlsx', index=False)
```

### Provide state ids to each state/ut as per state.csv


```python
mapping = {'ANDAMAN & NICOBAR ISLANDS':1, 'ANDHRA PRADESH':2, 'ARUNACHAL PRADESH' : 3, 'ASSAM': 4, 'BIHAR': 5,'CHHATTISGARH': 6,'DADRA & NAGAR HAVELI & DAMAN & DIU': 7,'DELHI': 8,'GOA': 9,
           'GUJARAT': 10,'HARYANA': 11,'HIMACHAL PRADESH': 12,'JHARKHAND': 13,'KARNATAKA': 14,'KERALA': 15,'LAKSHADWEEP':16,'MADHYA PRADESH':17, 'MAHARASHTRA': 18,'MANIPUR': 19,
           'MEGHALAYA': 20,'MIZORAM': 21,'NAGALAND': 22,'ODISHA': 23,'PUDUCHERRY': 24,'PUNJAB': 25,'RAJASTHAN': 26,'SIKKIM': 27,'TAMIL NADU': 28,'TELANGANA': 29,'TRIPURA': 30,'CHANDIGARH': 31,
           'JAMMU & KASHMIR': 32,'LADAKH': 33,'UTTAR PRADESH': 34,'UTTARAKHAND': 35,'WEST BENGAL': 36}
const = 0
for index, row in cen.iterrows():
    if row['district'] in mapping:
        const = mapping[row['district']]
    cen.at[index, 'state_id'] = const

cen.to_excel('2011-IndiaStateDist-0000.xlsx', index=False)
cen
```

### Add latitude and longitude for each district

```python
from geopy.geocoders import Nominatim

loc = Nominatim(user_agent="Geopy Library")

location_coords = []

cen['latitude'] = np.nan
cen['longitude'] = np.nan
```

```python
import geopy.exc

for index, row in cen.iterrows():
    print(index)
    if pd.isna(row['latitude']) == False:
        print('yes')
        continue
    
    getLoc = loc.geocode(row['district'])

    if getLoc == None:
        continue
    cen.at[index, 'latitude'] = getLoc.latitude
    cen.at[index, 'longitude'] = getLoc.longitude
    location_coords.append([getLoc.latitude, getLoc.longitude])
    print(row['district'], ":" , getLoc.latitude, getLoc.longitude)
    
```
