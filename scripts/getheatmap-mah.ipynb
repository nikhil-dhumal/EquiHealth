```python
# importing geopy library and Nominatim class
from geopy.geocoders import Nominatim
import folium
from folium.plugins import HeatMap
import pandas as pd
```

```python
MAHARASHTRA_COORDINATES = [19.75, 75.71]
# create a map centered at Maharashtra
map = folium.Map(location = MAHARASHTRA_COORDINATES, zoom_start=8)

loc = Nominatim(user_agent="Geopy Library")

location = loc.geocode("Maharashtra, India")

heat = []
```

```python
df2 = []

df = pd.read_csv('EquiHealth/data/maharashtra/hospital.csv')
i = -1
for dst in df['district']:
    i += 1
    getLoc = loc.geocode(dst)
    if getLoc == None:
        continue
    print(getLoc.address)
    heat.append([getLoc.latitude, getLoc.longitude])
    df2.append([df['hospital_id'][i], getLoc.latitude, getLoc.longitude])

HeatMap(heat).add_to(map)
map.save("maharashtra_map.html")
```

```python
x = pd.DataFrame(df2, columns=["hospital_id", "latitude", "longitude"])
x.to_csv("mah-hosp-lat-long.csv", index=False)
```
