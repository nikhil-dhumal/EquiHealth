Compute the latitudes and longitudes of each hospital using library geopy

```python
from geopy.geocoders import Nominatim
import pandas as pd
```

```python
df = pd.read_csv('hospital_directory.csv')
df
```

```python
loc = Nominatim(user_agent="Geopy Library")
```

```python
location_coords = []
```

```python
for index, row in df.iterrows():
    if pd.isna(row['Location_Coordinates']):
        getLoc = loc.geocode(row['District'])
        if getLoc == None:
            continue
        val = str(getLoc.latitude) + ", " + str(getLoc.longitude)
        df.at[index, 'Location_Coordinates'] = val
        location_coords.append(val)
        print(row['Hospital_Name'], ":" , val)
    
```

```python
df.to_csv('hospital_directory2.csv', index = False)
```
